```python
counts = df['location'].value_counts()
df2 = df.groupby('location')['location'].agg(count='count')
df2 = df2.loc[df2['count'] > 6000].reset_index()
df2.sort_values(by=['count'])
c = df2['location'].unique()
c.shape[0]
```
